```python
import numpy as np
hr_data = np.load("tm/HR_data_test_40.npy")
print("HR 数据 shape:", hr_data.shape)

idx = 0
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 加载 CSV
df = pd.read_csv("prediction_results_cGAN_Spring.csv")

# 提取数据并 reshape 为 (1400, 64, 64)
H, W = 64, 64
N = 100

# 反转为c
df['y_pred'] = df['y_pred'] - 273.15
df['y_true'] = df['y_true'] - 273.15

y_pred = df['y_pred'].values.reshape(N, H, W)
y_true = df['y_true'].values.reshape(N, H, W)
abs_error = df['abs_error'].values.reshape(N, H, W)

# 选择一个样本（如第0个）进行可视化

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
im0 = axs[0].imshow(y_true[idx], cmap='jet')
axs[0].set_title("Ground Truth")
plt.colorbar(im0, ax=axs[0], fraction=0.046)

im1 = axs[1].imshow(y_pred[idx], cmap='jet')
axs[1].set_title("Prediction")
plt.colorbar(im1, ax=axs[1], fraction=0.046)

im2 = axs[2].imshow(abs_error[idx], cmap='inferno')
axs[2].set_title("Absolute Error")
plt.colorbar(im2, ax=axs[2], fraction=0.046)

plt.suptitle(f"Sample #{idx} - Temperature Prediction")
plt.tight_layout()
plt.show()
```

```python
# === 提取中心行索引 ===
center_row = H // 2  # 取第32行（从0开始）

# === 提取中心行的温度数据 ===
true_profile = y_true[idx, center_row, :]
pred_profile = y_pred[idx, center_row, :]
error_profile = abs_error[idx, center_row, :]

# === 绘图 ===
plt.figure(figsize=(10, 5))
plt.plot(true_profile, label='Ground Truth', color='black', linewidth=2)
plt.plot(pred_profile, label='Prediction', color='blue', linestyle='--')
plt.plot(error_profile, label='Absolute Error', color='red', linestyle=':')
plt.title(f"Center Row Temperature Profile (Row {center_row}) - Sample #{idx}")
plt.xlabel("Column Index")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 保存第 idx 个样本的预测温度场为 CSV
output_array = y_pred[idx]  # shape: (H, W)
output_df = pd.DataFrame(output_array)

# 保存为 CSV
output_df.to_csv(f"prediction_sample_cGAN_Spring.csv", index=False, header=False)
print(f"✅ 已保存预测结果到 prediction_sample_cGAN_Spring.csv")

```

```python
# 保存第 idx 个样本的真实温度场为 CSV
real_array = y_true[idx]  # shape: (H, W)
real_df = pd.DataFrame(real_array)

# 保存为 CSV
real_df.to_csv("real_sample.csv", index=False, header=False)
print("已保存真实解到 real_sample.csv")
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import peak_signal_noise_ratio as psnr

# 读取数据
df = pd.read_csv("prediction_results_bicubic_winter.csv")
y_true = df["y_true"].values
y_pred = df["y_pred"].values

# 计算 MAE 和 RMSE
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

# reshape 为 (N, H, W)
N, H, W = 100, 64, 64
y_true_imgs = y_true.reshape(N, H, W)
y_pred_imgs = y_pred.reshape(N, H, W)

# 计算 SSIM 和 PSNR
ssim_scores = []
psnr_scores = []
for i in range(N):
    ssim_scores.append(ssim(y_true_imgs[i], y_pred_imgs[i], data_range=y_true_imgs[i].ptp()))
    psnr_scores.append(psnr(y_true_imgs[i], y_pred_imgs[i], data_range=y_true_imgs[i].ptp()))

print("MAE:", mae)
print("RMSE:", rmse)
print("SSIM:", np.mean(ssim_scores))
print("PSNR:", np.mean(psnr_scores))
```

# 综合画图

```python
import pandas as pd
import matplotlib.pyplot as plt

# === 加载三个模型的预测结果（64x64）===
esrgan = pd.read_csv("prediction_sample_ESRGAN.csv", header=None).values
srgan = pd.read_csv("prediction_sample_SRGAN.csv", header=None).values
cgan = pd.read_csv("prediction_sample_cGAN.csv", header=None).values
bilinear = pd.read_csv("prediction_sample_bilinear.csv", header=None).values
bicubic = pd.read_csv("prediction_sample_bicubic.csv", header=None).values
real = pd.read_csv("real_sample.csv", header=None).values

# === 提取中心行（第32行，索引31）===
center_idx = 31
esrgan_profile = esrgan[center_idx]
srgan_profile = srgan[center_idx]
cgan_profile = cgan[center_idx]
real_profile = real[center_idx]
bilinear_profile = bilinear[center_idx]
bicubic_profile = bicubic[center_idx]


# === 绘图 ===
plt.figure(figsize=(10, 5))
plt.plot(esrgan_profile, label='ESRGAN', color='blue', linestyle='-', marker='o', linewidth=2, markersize=4)
plt.plot(srgan_profile, label='SRGAN', color='green', linestyle='--', marker='s', linewidth=2, markersize=4)
plt.plot(cgan_profile, label='cGAN', color='red', linestyle='-.', marker='^', linewidth=2, markersize=4)
plt.plot(real_profile, label='Real', color='black', linestyle=':', marker='x', linewidth=2, markersize=4)
plt.plot(bilinear_profile, label='Bilinear', color='purple', linestyle='-', marker='D', linewidth=2, markersize=4)
plt.plot(bicubic_profile, label='Bicubic', color='orange', linestyle='--', marker='v', linewidth=2, markersize=4)
plt.title("Center Row Temperature Profile Comparison")
plt.xlabel("Column Index (centre)")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 直接比较两个模型预测差异
pred_esrgan = pd.read_csv("prediction_sample_ESRGAN.csv", header=None).values
pred_cgan = pd.read_csv("prediction_sample_cGAN.csv", header=None).values

diff = np.abs(pred_esrgan - pred_cgan)

# 可视化差异图
plt.imshow(diff, cmap='viridis')
plt.title("Difference between ESRGAN and cGAN Predictions")
plt.colorbar(label="|ESRGAN - cGAN| (°C)")
plt.tight_layout()
plt.show()

# 打印统计信息
print("Difference max:", np.max(diff))
print("Difference mean:", np.mean(diff))
```
